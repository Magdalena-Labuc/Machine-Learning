```python
#załadowanie bibliotek

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

Wykonaj następujące polecenia:

1. załaduj dane z pliku `train.csv` do dataframe

```python
df = pd.read_csv('train.csv') # wczytanie bazy danych
df
```

2. analogicznie do ćwiczeń z zajęć stwórz nowe zmienne wykonując następujące transformacje:

- `MSZoning` - one hot encoding 

```python
MSZoning = np.array([df['MSZoning']],ndmin=2).transpose()
encoder = OneHotEncoder(sparse=False)
encoder.fit(MSZoning)
MSZ_OHE = encoder.transform(MSZoning)
print(MSZ_OHE, ',\n', encoder.categories_)
```

- `LotShape` - one hot encoding

```python
LotShape = np.array([df['LotShape']],ndmin=2).transpose()
encoder = OneHotEncoder(sparse=False)
encoder.fit(LotShape)
LS_OHE = encoder.transform(LotShape)
print(LS_OHE, ',\n', encoder.categories_)
```

- `LotArea` - skalowanie od 0 do 1

```python
LotArea = np.array([df['LotArea']],ndmin=2).transpose()
scaler = MinMaxScaler()
scaler.fit(LotArea)
LA_scal = scaler.transform(LotArea)
print(scaler.data_min_, scaler.data_max_,',\n', LA_scal)
```

- `OverallCond` - standaryzacja

```python
OverallCond = np.array([df['OverallCond']],ndmin=2).transpose()
scaler = StandardScaler()
scaler.fit(OverallCond)
OC_scaler = scaler.transform(OverallCond)
print(OC_scaler)
```

3. Utworzone nowe zmienne dołącz do istniejącego zbioru (dataframe'u z pkt 1), w nazwie nowych kolumn po one hot encoding umieść kategorię której odpowiada:     
Przykładowo dla zmiennej `LotShape` powstaną 4 nowe kolumny o nazwach: `'LotShape_IR1'`, `'LotShape_IR2'`, `'LotShape_IR3'`, `'LotShape_Reg'`. 

```python
MSZ_df = pd.DataFrame(MSZ_OHE, columns=['MSZoning','MSZoning_FV','MSZoning_RH','MSZoning_RL','MSZoning_RM'])
LS_df = pd.DataFrame(LS_OHE, columns=['LotShape_IR1','LotShape_IR2','LotShape_IR3','LotShape_Reg'])
LA_df = pd.DataFrame(LA_scal, columns=['LotArea_scal'])
OC_df = pd.DataFrame(LA_scal, columns=['OC_scal'])
df2 = pd.concat([df, MSZ_df, LS_df, LA_df, OC_df], axis=1)
```
