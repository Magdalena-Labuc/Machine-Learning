Na podstawie zbioru `heart` zawierającego dane medyczne pacjentów wraz z informacją czy chorowali na serce (zmienna `target`: 1 - tak, 0 - nie) stwórz model drzewa decyzyjnego oraz lasu losowego, który będzie przewidywał ryzyko choroby.    

Opis poszczególnych zmiennych:     
- `age`: age in years    
- `sex`: sex (1 = male; 0 = female)
- `cp`: chest pain type    
    Value 1: typical angina     
    Value 2: atypical angina    
    Value 3: non-anginal pain    
    Value 4: asymptomatic    
- `trestbps`: resting blood pressure (in mm Hg on admission to the hospital)
- `chol`: serum cholestoral in mg/dl 
- `fbs`: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
- `restecg`: resting electrocardiographic results     
    Value 0: normal    
    Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)     
    Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
- `thalach`: maximum heart rate achieved 
- `exang`: exercise induced angina (1 = yes; 0 = no)
- `oldpeak` = ST depression induced by exercise relative to rest
- `slope`: the slope of the peak exercise ST segment    
    Value 1: upsloping    
    Value 2: flat    
    Value 3: downsloping     
- `ca`: number of major vessels (0-3) colored by flourosopy
- `thal`: 3 = normal; 6 = fixed defect; 7 = reversable defect

```python
import pandas as pd
import numpy as np
import seaborn as sns

import sklearn
from sklearn import tree
from sklearn import model_selection
import matplotlib.pyplot as plt
from mlxtend import plotting

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
```

```python
dane = pd.read_csv("heart.csv") #wczytanie danych
dane.head(10) #sprawdzenie pierwszych 10 linii
```

```python
dane.target.value_counts() #sprawdzenie target
```

```python
dane.describe() #sprawdzenie podstawowych parametrów statystycznych
```

```python
dane.info(10) # sprawdzenie podstawowych informacji, w tym null
```

```python
sns.countplot(x='sex', data=dane, palette="pastel") #policzenie liczby kobiet i mężczyzn
plt.xlabel("Sex (0 = female, 1= male)")
plt.show()
```

```python
countFemale = len(dane[dane.sex == 0])
countMale = len(dane[dane.sex == 1])
print("Ilość procentowa kobiet: {:.2f}%".format((countFemale / (len(dane.sex))*100)))
print("Ilość procentowa mężczyzn: {:.2f}%".format((countMale / (len(dane.sex))*100)))
```

```python
dane.groupby('target').mean()
```

```python
pd.crosstab(dane.age,dane.target).plot(kind="bar",figsize=(16,8))
plt.title('Ilość zawałów serca w rozróżnieniu na target czy chorowali wcześniej na serce (1 - tak, 0 - nie) w danym wieku')
plt.xlabel('Wiek')
plt.ylabel('Ilość wystąpień')
plt.show()
```

```python
x,y = plt.subplots(figsize=(12,12))
CorrelationMatrix = dane.drop("target",axis=1).corr()
sns.heatmap(CorrelationMatrix, vmin=-1.,vmax=1, annot=True)
plt.show()
```

```python
y = dane.target.values #usunięcie targetu z danych
x_data = dane.drop(['target'], axis = 1)
```

```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values #normalizacja danych
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

```python
#Drzewo decyzyjne

accuracies = {}

dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)

acc = dtc.score(x_test, y_test)*100
accuracies['Drzewo decyzyjne'] = acc
print("Wynik testu dokładności drzewa decyzyjnego {:.2f}%".format(acc))
```

```python
#Las losowy
rf = RandomForestClassifier(n_estimators = 1000, random_state = 1)
rf.fit(x_train, y_train)

acc = rf.score(x_test,y_test)*100
accuracies['Las losowy'] = acc
print("Wynik testu dokładności lasu losowego : {:.2f}%".format(acc))
```

```python
#porównanie dwóch modeli na wykresie

colors = ["yellow", "magenta"]

sns.set_style("whitegrid")
plt.figure(figsize=(16,5))
plt.yticks(np.arange(0,100,10))
plt.ylabel("Accuracy %")
plt.xlabel("Algorithms")
sns.barplot(x=list(accuracies.keys()), y=list(accuracies.values()), palette=colors)
plt.show()
```

# Wniosek:

Las losowy jest lepszy pod względem dokładności.

Krzywa ROC

```python
probs_from_dtc = dtc.predict_proba(x_test)[:, 1]
probs_from_rf = rf.predict_proba(x_test)[:, 1]

from sklearn.metrics import roc_curve

dt_fpr, dt_tpr, dt_thresholds = roc_curve(y_test, probs_from_dtc)
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, probs_from_rf)


from matplotlib import pyplot

pyplot.plot(dt_fpr, dt_tpr, marker='.', label='Drzewo decyzyjne')
pyplot.plot(rf_fpr, rf_tpr, marker='.', label='Las losowy')
# axis labels
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
# show the legend
pyplot.legend()
# show the plot
pyplot.show()
```

# Wniosek:
Im bardziej wykres zmierza ku lewej górnej części wykresu, czyli do punktu (0,1) i im AUC, czyli 
pole powierzchni pod wykresem ROC jest większe, tym jest lepszy wykres.

Porównując dwa modele, czyli drzewo decyzyjne i las losowy, wybrałabym las losowy ponieważ wykres ten będzie dokładniej
przewidywał ryzyko choroby.
